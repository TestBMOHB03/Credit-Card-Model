```python
y_test_dd = dd.read_csv('y_test_2')
y_test_dd.head()
```

```python
classifier = skflow.TensorFlowDNNClassifier(
  hidden_units=[10], n_classes=3, steps=10000, learning_rate=0.1, tf_random_seed=9)
classifier.fit(X_train_dd, y_train_dd)
```

```python
y_pred = X_test_dd.map_partitions(classifier.predict).compute()
accuracy_score(y_test_dd.compute(), y_pred)
```
